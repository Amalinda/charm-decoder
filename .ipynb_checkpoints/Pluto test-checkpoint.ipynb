```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
```

```python
samples = np.fromfile("data/915MHz_2048kHz_8dB_cap", dtype = complex, count = -1)
print(samples.shape)
```

```python
fs = 1024000
f, t, FFTsignal = spectrogram(samples[0:int(fs*0.2)], fs = fs, nperseg = 1024, noverlap = 512, mode = 'complex', return_onesided = False)

print(t.shape, f.shape, FFTsignal.shape)

plt.figure()
plt.pcolormesh(t, f, np.abs(FFTsignal))
plt.ylabel("Freq (Hz)")
plt.xlabel("Time (s)")
plt.show()
```

```python
f, t, FFTsignal = spectrogram(samples[int(fs*0.025):int(fs*0.065)], fs = fs, nperseg = 128, noverlap = 64, mode = 'psd', return_onesided = False)
# TODO: might have to add time offset

t = t + 0.025

plt.figure(figsize = (20,3))
plt.pcolormesh(t, f, FFTsignal)
plt.ylabel("Freq (Hz)")
plt.xlabel("Time (s)")
plt.show()
```
