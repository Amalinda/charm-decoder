```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
filename = "data/915MHz_2048kHz_8dB_cap"

samples = np.fromfile(filename, dtype = complex, count = -1)
print(samples.shape)

fs = 2048000

# teh spectrogram signal is wrong (basedon comparison with the MTALB spectrogram)
f, t, FFTsignal = signal.spectrogram(samples[0:int(fs*0.2)], fs = fs, nperseg = 1024, noverlap = 512, mode = 'psd', return_onesided = False)

print(t.shape, f.shape, FFTsignal.shape)

plt.figure()
plt.pcolormesh(t, f, (FFTsignal))
plt.ylabel("Freq (Hz)")
plt.xlabel("Time (s)")
plt.show()
```
